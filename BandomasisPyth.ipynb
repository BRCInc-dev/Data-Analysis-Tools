```python
import statsmodels.formula.api as smf

# Sukurkia modelį su TEMP kaip prognozavimo kintamuoju
model1 = smf.ols('PRES ~ TEMP', data=data).fit()

# Išsaugo R-kvadratą
r_squared1 = model1.rsquared

# Pateikiu modelio santrauką
print(model1.summary())
print(f"R-squared for model 1: {r_squared1}")

# Sukurkia modelį su keliais prognozavimo kintamaisiais (pvz., TEMP ir DEWP)
model2 = smf.ols('PRES ~ TEMP + DEWP', data=data).fit()  # Pavyzdys su TEMP ir DEWP

# Išsaugo R-kvadratą
r_squared2 = model2.rsquared  # Priskiriama reikšmė r_squared2

print(f"R-squared for model 1 (simple linear regression): {r_squared1}")
print(f"R-squared for model 2 (multiple linear regression): {r_squared2}")

if r_squared2 > r_squared1:
    print("Multiple linear regression model performs better.")
else:
    print("Simple linear regression model performs better.")
```
