```python
!pip install pandas matplotlib scikit-learn imbalanced-learn xgboost seaborn

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
import numpy as np
```

```python
# Load the dataset
data_path = "/content/customer_churn_dataset-training-master (4).csv"
data = pd.read_csv(data_path)
```

```python
# Separate features and target variable
# Check if 'Churn' column has any valid values other than NaN
if data['Churn'].dropna().empty:
    print("Error: 'Churn' column is empty or contains only NaN values.")
else:
    print("'Churn' column has valid data.")
data['Churn'] = data['Churn'].map({'Yes': 1, 'No': 0})
X = data.drop('Churn', axis=1)
y = data['Churn']
```

```python
# --- Preprocessing ---
# Identify categorical columns
categorical_cols = X.select_dtypes(include=['object']).columns
```

```python
# Create a column transformer for one-hot encoding
preprocessor = ColumnTransformer(
    transformers=[
        ('onehot', OneHotEncoder(handle_unknown='ignore'), categorical_cols)],
    remainder='passthrough')
```

```python
# Fit and transform X
X_processed = preprocessor.fit_transform(X)
```

```python
# Convert back to dataframe for easier handling
feature_names = preprocessor.get_feature_names_out()
X_processed = pd.DataFrame(X_processed, columns=feature_names)
```

```python
# Handle any NaN values in X_processed by imputing with the mean
X_processed = X_processed.fillna(X_processed.mean())

# --- End of preprocessing ---
```

```python
# Split the data into training and testing sets
# Remove na values from y
if not y.dropna().empty: # Check if y has values before dropna
    y = y.dropna()
    X_processed = X_processed.iloc[y.index]
    X_train, X_test, y_train, y_test = train_test_split(X_processed, y, test_size=0.2, random_state=42)
else:
    print("Error: 'Churn' column is empty, no data to split.")
```

```python
# Initialize models
models = {
    'Logistic Regression': LogisticRegression(random_state=42),
    'Random Forest': RandomForestClassifier(random_state=42),
    'XGBoost': xgb.XGBClassifier(random_state=42)
}
if not y.dropna().empty:
    # Train the models
    for model_name, model in models.items():
        model.fit(X_train, y_train)
```
