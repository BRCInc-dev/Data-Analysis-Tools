```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Atsisiųskite duomenis
df = pd.read_excel('irisai.xlsx')
```

```python
df.describe()
```

```python
# Histograma
sns.histplot(df['Žiedlapio ilgis (cm)'])

# Pavadinimas
plt.title("Žiedlapių ilgis")
```

```python
# Histograma 2, pagal rūšį su spalvomis.
sns.histplot(data=df, x='Žiedlapio ilgis (cm)', hue='Irisų rūšis')

# Pavadinimas
plt.title("Žiedlapių ilgio histograma pagal rūšį")
```

```python
# KDE diagrama
sns.kdeplot(data=df, x='Žiedlapio ilgis (cm)', hue='Irisų rūšis', fill=True)

# Pavadinimai
plt.title("Žiedų ilgis pagal rūšį")
```

```python
# 2D KDE diagrama
sns.jointplot(x=df['Žiedlapio ilgis (cm)'], y=df['Taurėlapio plotis (cm)'], kind="kde")
```

```python
# 2D KDE diagrama
spalva = sns.jointplot(x=df['Žiedlapio ilgis (cm)'], y=df['Taurėlapio plotis (cm)'], kind="kde", fill=True, cmap="Blues")

plt.show()
```
